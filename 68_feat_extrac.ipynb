```python
from pyAudioAnalysis import audioBasicIO
from pyAudioAnalysis import ShortTermFeatures
import matplotlib.pyplot as plt
import numpy
import scipy.io.wavfile as wav
import os

# directory where we your .wav files are
directoryName = r"/home/user/Sahil/dataset" # put your own directory here
# directory to put our results in, you can change the name if you like
resultsDirectory = r"/home/user/Sahil/orig_features_68"

for filename in os.listdir(directoryName):
    
    if filename.endswith(".wav") : 
        [Fs, x] = audioBasicIO.read_audio_file(directoryName + "/" +filename)
        #print(filename)
        F, f_names = ShortTermFeatures.feature_extraction(x, Fs, 1024, 512)
        print(F.shape)
        # print(F)
        # print(f_names)
        # create a file to save our results in
        outputFile = resultsDirectory + "/" + os.path.splitext(filename)[0]+'.npy'
        numpy.save(outputFile,F) #save MFCCs as .csv

```
